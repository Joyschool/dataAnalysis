#### 배열의 크기 변경

```python
na = np.arange(12)
print(f'na : {na}')

nb = na.reshape(3, 4)
print(f'nb : {nb}')

nc = na.reshape(2, -1)
print(f'nc : {nc}')

nd = na.reshape(2, -1, 2)
print(f'nd : {nd}')
```

### 배열의 연산

```python
x = np.arange(1, 10001)
y = np.arange(10001, 20001)
z = x + y
z = np.add(x+y)
print(z)
```

```python
x = np.arange(1,21)
print(x)
```

```python
print(f'1번(3의 배수) 답: {x[x%3==0]}')
print(f'2번(4로 나누어 나머지가 1)답: {x[x%4==1]}')
print(f'3번(3의 배수이고 4로 나누어 나머지가 1) 답: {x[(x%3==0)&(x%4==1)]}')
```
